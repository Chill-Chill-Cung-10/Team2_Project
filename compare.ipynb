```python
import pandas as pd
```

```python
csv_1 = pd.read_csv("data_warehouse/sleep_quality_vs_student_performance_(2).csv")
csv_2 = pd.read_csv("data_warehouse/sleep_quality_vs_student_performance.csv")

cols1 = set(csv_1.columns)
cols2 = set(csv_2.columns)
csv_1
# print(len(cols1) == len(cols2))
```

```python
rows1 = set(tuple(r) for r in csv_1.itertuples(index=False, name=None))
rows2 = set(tuple(r) for r in csv_2.itertuples(index=False, name=None))
only_in_1 = rows1 - rows2
only_in_2 = rows2 - rows1
print("Rows only in csv_1:", len(only_in_1))
print("Rows only in csv_2:", len(only_in_2))

# Show up to 5 examples
for r in list(only_in_1)[:5]:
    print("Only in csv_1 example:", r)
for r in list(only_in_2)[:5]:
    print("Only in csv_2 example:", r)

# 4) So sánh theo khoá (nếu muốn): đặt key thành tên cột hoặc danh sách cột
key = None  # ví dụ: ['student_id'] hoặc ['student_id','date']
if key:
    merged = csv_1.merge(csv_2, on=key, how='outer', indicator=True, suffixes=('_1','_2'))
    print("Merge indicator counts:\n", merged['_merge'].value_counts())
    # show differences cho các cột chung (không tính key)
    common_nonkey = (set(csv_1.columns) & set(csv_2.columns)) - set(key)
    for col in common_nonkey:
        a = f"{col}_1"; b = f"{col}_2"
        diff = merged[(merged[a].notna() | merged[b].notna()) & (merged[a] != merged[b])]
        if not diff.empty:
            print(f"Differences in column {col}:")
            print(diff[[*key, a, b]].head())
```
