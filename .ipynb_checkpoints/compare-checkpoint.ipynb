```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import warnings
from sklearn.model_selection import train_test_split
warnings.filterwarnings('ignore')
# Cấu hình để biểu đồ hiển thị ngay trong notebook
%matplotlib inline
```

```python
csv_1 = pd.read_csv("data_warehouse/sleep_to_performance.csv")
csv_2 = pd.read_csv("data_warehouse/habit_to_performance.csv")

cols1 = set(csv_1.columns)
cols2 = set(csv_2.columns)
csv_1
# print(len(cols1) == len(cols2))
```

```python
rows1 = set(tuple(r) for r in csv_1.itertuples(index=False, name=None))
rows2 = set(tuple(r) for r in csv_2.itertuples(index=False, name=None))
only_in_1 = rows1 - rows2
only_in_2 = rows2 - rows1
print("Rows only in csv_1:", len(only_in_1))
print("Rows only in csv_2:", len(only_in_2))

# Show up to 5 examples
for r in list(only_in_1)[:5]:
    print("Only in csv_1 example:", r)
for r in list(only_in_2)[:5]:
    print("Only in csv_2 example:", r)

# 4) So sánh theo khoá (nếu muốn): đặt key thành tên cột hoặc danh sách cột
key = None  # ví dụ: ['student_id'] hoặc ['student_id','date']
if key:
    merged = csv_1.merge(csv_2, on=key, how='outer', indicator=True, suffixes=('_1','_2'))
    print("Merge indicator counts:\n", merged['_merge'].value_counts())
    # show differences cho các cột chung (không tính key)
    common_nonkey = (set(csv_1.columns) & set(csv_2.columns)) - set(key)
    for col in common_nonkey:
        a = f"{col}_1"; b = f"{col}_2"
        diff = merged[(merged[a].notna() | merged[b].notna()) & (merged[a] != merged[b])]
        if not diff.empty:
            print(f"Differences in column {col}:")
            print(diff[[*key, a, b]].head())
```

```python
# %%
# Tải bộ dữ liệu vào biến 'csv_3'
file_path = 'data_warehouse/other_factors.csv'
csv_3 = pd.read_csv(file_path)

# Hiển thị thông tin cơ bản
print("--- Thông tin Dữ liệu (Data Info) ---")
csv_3.info()

# Hiển thị 5 hàng đầu tiên
print("\n--- 5 hàng dữ liệu đầu tiên (Data Head) ---")
print(csv_3.head())

# Kiểm tra các giá trị bị thiếu
print("\n--- Tổng số giá trị bị thiếu (Missing Values) ---")
print(csv_3.isnull().sum())

# Đếm giá trị cho các cột phân loại quan trọng
print("\n--- Phân bố 'School_Type' ---")
print(csv_3['School_Type'].value_counts())

print("\n--- Phân bố 'Gender' ---")
print(csv_3['Gender'].value_counts())

print("\n--- Phân bố 'Parental_Education_Level' ---")
print(csv_3['Parental_Education_Level'].value_counts())
```

```python
## 2. Trực quan hóa Dữ liệu
# --- Biểu đồ 1: Phân bố của Điểm Thi (Exam_Score) ---
plt.figure(figsize=(10, 6))
plt.hist(csv_3['Exam_Score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Phân bố của Điểm Thi (Exam_Score)', fontsize=16)
plt.xlabel('Điểm Thi', fontsize=12)
plt.ylabel('Tần suất (Số lượng học sinh)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.tight_layout()
plt.show()
```

```python
# --- Biểu đồ 2: Mối quan hệ giữa Số Giờ Học và Điểm Thi ---
plt.figure(figsize=(10, 6))
plt.scatter(csv_3['Hours_Studied'], csv_3['Exam_Score'], alpha=0.5, color='purple')
plt.title('Mối quan hệ giữa Số Giờ Học và Điểm Thi', fontsize=16)
plt.xlabel('Số Giờ Học (Hours_Studied)', fontsize=12)
plt.ylabel('Điểm Thi (Exam_Score)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# --- Biểu đồ 3: Phân bố Điểm Thi theo Giới tính ---

# Chuẩn bị dữ liệu
male_scores = csv_3[csv_3['Gender'] == 'Male']['Exam_Score']
female_scores = csv_3[csv_3['Gender'] == 'Female']['Exam_Score']

plt.figure(figsize=(8, 6))
# Tạo boxplot
bp = plt.boxplot([male_scores, female_scores], patch_artist=True, labels=['Nam (Male)', 'Nữ (Female)'])

# Tô màu
colors = ['lightblue', 'lightpink']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    
plt.title('Phân bố Điểm Thi theo Giới tính', fontsize=16)
plt.xlabel('Giới tính', fontsize=12)
plt.ylabel('Điểm Thi', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# --- Biểu đồ 4: Phân bố Điểm Thi theo Trình độ Học vấn của Phụ huynh ---

# Xử lý dữ liệu: Bỏ qua các hàng bị thiếu 'Parental_Education_Level'
df_edu_cleaned = csv_3.dropna(subset=['Parental_Education_Level'])

# Xác định thứ tự
edu_order = ['High School', 'College', 'Postgraduate']

# Chuẩn bị dữ liệu theo thứ tự
scores_by_edu = [df_edu_cleaned[df_edu_cleaned['Parental_Education_Level'] == level]['Exam_Score'] for level in edu_order]

plt.figure(figsize=(12, 7))
bp_edu = plt.boxplot(scores_by_edu, patch_artist=True, labels=edu_order)

# Tô màu
colors = ['#FFC3A0', '#FF677D', '#D4A5A5']
for patch, color in zip(bp_edu['boxes'], colors):
    patch.set_facecolor(color)
    
plt.title('Phân bố Điểm Thi theo Trình độ Học vấn của Phụ huynh', fontsize=16)
plt.xlabel('Trình độ Học vấn Phụ huynh', fontsize=12)
plt.ylabel('Điểm Thi', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
